```python
def convert_to_gray(frame):
    return cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

def apply_gaussian_filter(frame):
    return cv2.GaussianBlur(frame, (5, 5), 0)

def process_video(video_path, processing_function, results, lock, start_time):
    cap = cv2.VideoCapture(video_path)

    while True:
        ret, frame = cap.read()
        frame = cv2.resize(frame,(550,500))
        if not ret:
            break

        processed_frame = processing_function(frame)

        with lock:
            results.append((processed_frame, time.time() - start_time))

    cap.release()

if __name__ == "/kaggle/working/":
    video_path_gray = "/kaggle/input/video-test/video1.mp4"
    video_path_gaussian = "/kaggle/input/video-test/video2.mp4"
    
    frames_gray = []
    frames_gaussian = []
    results_lock = threading.Lock()

    # Record start time for each thread
    start_time_gray = time.time()
    start_time_gaussian = time.time()

    # Create threads for each video processing
    thread_gray = threading.Thread(target=process_video, args=(video_path_gray, convert_to_gray, frames_gray, results_lock, start_time_gray))
    thread_gaussian = threading.Thread(target=process_video, args=(video_path_gaussian, apply_gaussian_filter, frames_gaussian, results_lock, start_time_gaussian))

    # Start both threads
    thread_gray.start()
    thread_gaussian.start()

    # Main thread continues to display processed frames
    while True:
        with results_lock:
            if frames_gray and frames_gaussian:
                # Display or save the processed frames for both videos
                cv2.imshow("Gray Video", frames_gray.pop(0)[0])
                cv2.imshow("Gaussian Video", frames_gaussian.pop(0)[0])

                # Print start time for each thread
                print("Start Time - Gray Thread:", start_time_gray)
                print("Start Time - Gaussian Thread:", start_time_gaussian)

                cv2.waitKey(30)  # Adjust the waitKey value based on your video frame rate

        # Add other main thread tasks here

    # Wait for both threads to finish
    thread_gray.join()
    thread_gaussian.join()

    cv2.destroyAllWindows()
```
